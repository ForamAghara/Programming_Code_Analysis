#For 1 Author

```python
# Reading the metadata file for author 1 label's data

files_list1 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Dataset/1/metadata-1.xlsx")
files_list1
```

```python
# Calculate time cost in days from first commit and last commit

import datetime

TIME_COST_DAYS = []

for index in range(len(files_list1)):
  start = files_list1.iloc[index]['First_commit']
  end = files_list1.iloc[index]['Last_commit']
  
  #Extracting the date in date format from the string date stored
  format = '%b %d, %Y' 
  date_str = datetime.datetime.strptime(start, format).date()
  date_end = datetime.datetime.strptime(end, format).date()

  #Getting Absolute value of the difference of the days
  difference = abs(date_end-date_str).days

  TIME_COST_DAYS.append(difference)

# Adding the time cost difference to the metadata file 

files_list1['TIME_COST_IN_DAYS'] = TIME_COST_DAYS
```

```python
# Creating the empty list to store the values of number of lines of code, tokens, functions, complexity and parameters from the source codes 
NLOC_1 = []
TOKENS_1 = []
FUNCTIONS_1 = []
COMPLEXITY_1 = []
PARAMS_1 = []


for i in files_list1['Filename']:
  full_loc = "/content/drive/MyDrive/Colab Notebooks/Dataset/1/files/"+i

  # analysing the programming code files using lizard library
  key_features = lizard.analyze_file(full_loc)

  functions = len(key_features.function_list)

  # storing the information extracted by the lizard library
  NLOC_1.append(key_features.__dict__['nloc'])
  TOKENS_1.append(key_features.__dict__['token_count'])
  FUNCTIONS_1.append(functions)

  #Summation of the cyclomatic complexity and total parameters from the each function present in the file
  total_ccn = 0
  total_params = 0
  for func in range(0,functions):
      total_ccn += key_features.function_list[func].__dict__['cyclomatic_complexity']
      total_params += len(key_features.function_list[func].__dict__['full_parameters'])

  COMPLEXITY_1.append(total_ccn)
  PARAMS_1.append(total_params)

# Adding the number of lines of code, tokens, functions, complexity and parameters to the metadata file 

files_list1['NLOC'] = NLOC_1
files_list1['TOTAL TOKENS'] = TOKENS_1
files_list1['TOTAL FUNCTIONS'] = FUNCTIONS_1
files_list1['CYCLOMATIC COMPLEXITY'] = COMPLEXITY_1
files_list1['TOTAL PARAMETERS'] = PARAMS_1
```

```python
#Formats to look up for "if", "switch", "for" and "while" statements in the source codes 

if_state = ["if (", "if("]
switch_state = ["switch (", "switch("]
for_state = ["for (", "for("]
while_state = ["while (", "while("]
```

```python
# Creating the empty list to store the counts of number of "if", "switch", "for" and "while" statements in the source codes  

IF_1 = []
FOR_1 = []
WHILE_1 = []
SWITCH_1 = []

for i in files_list1['Filename']:
  full_loc = "/content/drive/MyDrive/Colab Notebooks/Dataset/1/files/"+i

  with open(full_loc) as f:
    src = f.read()

  cntif = 0
  cntswitch = 0
  cntfor = 0
  cntwhile = 0

  # counting the "if", "switch", "for" and "while" statements
  for i in range(len(if_state)):
    cntif += src.count(if_state[i])
  for i in range(len(switch_state)):
    cntswitch += src.count(switch_state[i])
  for i in range(len(for_state)):
    cntfor += src.count(for_state[i])
  for i in range(len(while_state)):
    cntwhile += src.count(while_state[i])

  IF_1.append(cntif)
  FOR_1.append(cntfor)
  WHILE_1.append(cntwhile)
  SWITCH_1.append(cntswitch)

# Adding the number of "if", "switch", "for" and "while" statements to the metadata file

files_list1["Number of IF clause"] = IF_1
files_list1["Number of FOR clause"] = FOR_1
files_list1["Number of WHILE clause"] = WHILE_1
files_list1["Number of SWITCH clause"] = SWITCH_1
```

#MERGING DATASET

```python
#Merging the feature extracted metadata files to create whole processed dataset

data = [files_list1,files_list2,files_list3,files_list4,files_list5]

merged_dataset = pd.concat(data,ignore_index=True)
```

```python
merged_dataset
```

```python
# Creating list of authors label values and adding list to the final merged dataset
auth_list = []

for i in range(1,6):
  for j in range(40):
    auth_list.append(i)

merged_dataset["Number_of_Authors"] = auth_list
```

```python
merged_dataset
```

```python
# creating time class column with the values generated using thresholds of the time cost in days column

TIME_CLASS = []
for i in range(len(merged_dataset)):
  if (merged_dataset.iloc[i]["TIME_COST_IN_DAYS"]<=100):
    TIME_CLASS.append(1)
  elif ((merged_dataset.iloc[i]["TIME_COST_IN_DAYS"]>100) & (merged_dataset.iloc[i]["TIME_COST_IN_DAYS"]<=250)):
    TIME_CLASS.append(2)
  elif ((merged_dataset.iloc[i]["TIME_COST_IN_DAYS"]>250) & (merged_dataset.iloc[i]["TIME_COST_IN_DAYS"]<=500)):
    TIME_CLASS.append(3)
  elif ((merged_dataset.iloc[i]["TIME_COST_IN_DAYS"]>500) & (merged_dataset.iloc[i]["TIME_COST_IN_DAYS"]<=1000)):
    TIME_CLASS.append(4)
  elif ((merged_dataset.iloc[i]["TIME_COST_IN_DAYS"]>1000) & (merged_dataset.iloc[i]["TIME_COST_IN_DAYS"]<=2000)):
    TIME_CLASS.append(5)
  elif (merged_dataset.iloc[i]["TIME_COST_IN_DAYS"]>2000):
    TIME_CLASS.append(6)

merged_dataset["Time_Class"] = TIME_CLASS
```
