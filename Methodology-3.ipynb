# Importing libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing the merged dataset processed and remove the time class column because is will overfit with time cost in days feature

```python
dataset = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Dataset/merged_dataset_processed.csv")
code_analysis_data = dataset.drop(columns=["Time_Class"])
```

# Printing the correlation Matrix among code features

```python
plt.figure(figsize=(20,10))
corrMatrix = code_analysis_data.corr()
sns.heatmap(corrMatrix, annot=True)
plt.title("Correlation Matrix of Code Characteristics")
plt.show()
```

# Generating Scatter Plot matrix for time cost in days feature vs all other code features

```python
plt.figure(figsize=(20,20))
plt.subplot(4,3,1)
a=sns.regplot(x="TIME_COST_IN_DAYS", y="CYCLOMATIC COMPLEXITY", data=code_analysis_data)
a.set_xlim(0,1000)
plt.title("Correlation with Cyclomatic Complexity")
plt.subplot(4,3,2)
b=sns.regplot(x="TIME_COST_IN_DAYS", y="Number_of_Authors", data=code_analysis_data)
b.set_xlim(0,1000)
plt.title("Correlation with Number of Authors")
plt.subplot(4,3,3)
c=sns.regplot(x="TIME_COST_IN_DAYS", y="Total_commits", data=code_analysis_data)
c.set_xlim(0,1000)
plt.title("Correlation with Total Number of Commits")
plt.subplot(4,3,4)
d=sns.regplot(x="TIME_COST_IN_DAYS", y="NLOC", data=code_analysis_data)
d.set_xlim(0,1000)
plt.title("Correlation with Number of Lines of Code")
plt.subplot(4,3,5)
e=sns.regplot(x="TIME_COST_IN_DAYS", y="TOTAL TOKENS", data=code_analysis_data)
e.set_xlim(0,1000)
plt.title("Correlation with Total number of Tokens")
plt.subplot(4,3,6)
f=sns.regplot(x="TIME_COST_IN_DAYS", y="TOTAL FUNCTIONS", data=code_analysis_data)
f.set_xlim(0,1000)
plt.title("Correlation with Total number of Functions")
plt.subplot(4,3,7)
g= sns.regplot(x="TIME_COST_IN_DAYS", y="TOTAL PARAMETERS", data=code_analysis_data)
g.set_xlim(0,1000)
plt.title("Correlation with Total number of Parameters")
plt.subplot(4,3,8)
h=sns.regplot(x="TIME_COST_IN_DAYS", y="Number of IF clause", data=code_analysis_data)
h.set_xlim(0,1000)
plt.title("Correlation with Total number of IF clause")
plt.subplot(4,3,9)
i=sns.regplot(x="TIME_COST_IN_DAYS", y="Number of FOR clause", data=code_analysis_data)
i.set_xlim(0,1000)
plt.title("Correlation with Total number of FOR clause")
plt.subplot(4,3,10)
j=sns.regplot(x="TIME_COST_IN_DAYS", y="Number of WHILE clause", data=code_analysis_data)
j.set_xlim(0,1000)
plt.title("Correlation with Total number of WHILE clause")
plt.subplot(4,3,11)
k=sns.regplot(x="TIME_COST_IN_DAYS", y="Number of SWITCH clause", data=code_analysis_data)
k.set_xlim(0,1000)
plt.title("Correlation with Total number of SWITCH clause")
plt.show()
```
