# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib 
import seaborn as sns
```

# Reading the results of 4 cases developed for methodology 2

```python
#Reading the results of 4 cases developed for methodology 2

a1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Dataset/Results/M2(a.1).csv")
a2 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Dataset/Results/M2(a.2).csv")
b1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Dataset/Results/M2(b.1).csv")
b2 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Dataset/Results/M2(b.2).csv")
```

# Adding extra 2 columns to provide information on Cyclomatic Complexity Feature and Number of Authors features is included or not.

```python
#Creating 2 empty lists

author = []
features = []

# Adding values as "No" to the list where number of author feature is not present and also "Yes" to the list where number of author feature is present 
for i in range(2):
  for i in range(0,12):
    author.append("No")
for i in range(2):
  for i in range(0,12):
    author.append("Yes")

# Adding values as "No" to the list where cyclomatic complexity feature is not present and also "Yes" to the list where cyclomatic complexity feature is present
for i in range(0,12):
  features.append("No")
for i in range(0,12):
  features.append("Yes")
for i in range(0,12):
  features.append("No")
for i in range(0,12):
  features.append("Yes")

```

# Combining the results

```python
#Merging the results scores

Merged_Scores = pd.concat([a1,a2,b1,b2],ignore_index=True)
```

```python
#Adding the 2 columns with the list values created above

Merged_Scores["Author_Included"] = author
Merged_Scores["Cyclomatic_Complexity_Included"] = features
```

```python
#Printing the combined results

Merged_Scores
```

# Performance Evaluation of models when Number of Authors is not included

```python
#Performance Evaluation of models when Number of Authors is not included

no_auth_results = Merged_Scores[Merged_Scores["Author_Included"] == "No"]  
sns.catplot(x="Classifier", y="Accuracy", hue="Hyperparameter_Tuning", col="Cyclomatic_Complexity_Included",
                data=no_auth_results, kind="point",
                 height=8, aspect=.7);
```

# Performance Evaluation of models when Number of Authors is included

```python
#Performance Evaluation of models when Number of Authors is included

auth_results = Merged_Scores[Merged_Scores["Author_Included"] == "Yes"]  
sns.catplot(x="Classifier", y="Accuracy", hue="Hyperparameter_Tuning", col="Cyclomatic_Complexity_Included",
                data=auth_results, kind="point",
                 height=8, aspect=.7);
```

# Performance Evaluation of models when Cyclomatic Complexity is not included

```python
#Performance Evaluation of models when Cyclomatic Complexity is not included

oc_results = Merged_Scores[Merged_Scores["Cyclomatic_Complexity_Included"] == "No"]  
sns.catplot(x="Classifier", y="Accuracy", hue="Hyperparameter_Tuning", col="Author_Included",
                data=oc_results, kind="point",
                 height=8, aspect=.7);
```

# Performance Evaluation of models when Cyclomatic Complexity is included

```python
#Performance Evaluation of models when Cyclomatic Complexity is included

cc_results = Merged_Scores[Merged_Scores["Cyclomatic_Complexity_Included"] == "Yes"]  
sns.catplot(x="Classifier", y="Accuracy", hue="Hyperparameter_Tuning", col="Author_Included",
                data=cc_results, kind="point",
                 height=8, aspect=.7);
```

# Getting the Maximum accuracy when Cyclomatic complexity is not included

```python
#Getting the Maximum accuracy when Cyclomatic complexity is not included

no_cc = Merged_Scores[Merged_Scores["Cyclomatic_Complexity_Included"] == "No"]
print("Maximum Accuracy when Cyclomatic Complexity Not Included :", max(no_cc['Accuracy']*100),"%")  
```

# Getting the Maximum accuracy when Cyclomatic complexity is included

```python
#Getting the Maximum accuracy when Cyclomatic complexity is included

cc = Merged_Scores[Merged_Scores["Cyclomatic_Complexity_Included"] == "Yes"]
print("Maximum Accuracy when Cyclomatic Complexity Included :", max(cc['Accuracy']*100),"%") 
```
